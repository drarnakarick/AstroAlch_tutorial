## SQLAlchemy Goals and objectives

- Provide tools and components to assist database development.
- Allow expression of DB/SQL tasks using declarative patterns.
- Develop a consistent interface to use distinct features of each backend, with hiding them.

SQLAlchemy core contains:

- **Engine**: A registry which provides connectivity to a particular database server.
- **Dialect**: Interprets generic SQL commands in terms of specific database vocabulory
- **Connection Pool**: Holds collection of database connections in memory for fast resuse
- **SQL Expression Language**: Allows SQL statements to be written using python expressios
- **Schema**: Uses python objects to represent tables, columns and datatypes

ORM does the following:

- Allows construction of Python objects which can be mapped to relational database tables.
- Persists objects into their corresponding data tables.
- Provides a query statement to load objects using SQL generated from mappings
- Builds on top of the core

## An Inside Out View

There are various abstraction levels in SQLAlchemy and you can work on any of them.

* At first layer, we have `Engine` which maintains classical DBAPI interactions. DBAPI is the standard python database interface. Engine provides connectionless execution which means it connects and disconnects for us.
* At second layer, we have `Metadata` which describes the structure of database: table, columns and datatypes. Metadata can be used to generate a schema and the other way around: that is, it can be generated from a schema. At this layer, an SQL query can be explicity performed using engine.execute()
* At layer 3, we have `SQL Expressions` which builds upong metadata to compose SQL statements in Python. We use python objects to represent SQL expressions.
* At layer 4, we have `ORM` or Object Relational Mapping that associates object oriented classes with database tables.
